```python
import pandas as pd
```

```python
df_ventas = pd.read_csv('../dataset/procesado/ventas_limpio.csv')

df_ventas.head()
```

```python
# KPI principal (Ventas totales)
ventas_totales = df_ventas['TotalVenta'].sum()

print(f'Venta total {ventas_totales:,.2f}')
```

```python
# Top 10 ventas por productos
top_ventas_productos = df_ventas.groupby('NombreProducto', as_index = False).agg(
    TotalVenta = ('TotalVenta', 'sum')).sort_values('TotalVenta', ascending = False).head(10)

top_ventas_productos
```

```python
# Top 5 clientes que más compra
top_cliente = df_ventas.groupby('clienteID', as_index = False).agg(
    TotalVenta = ('TotalVenta', 'sum')).sort_values('TotalVenta', ascending = False).head(5)

top_cliente
```

```python
# Venta por mes
venta_mes = df_ventas.groupby('Mes', as_index = False).agg(
    TotalVenta = ('TotalVenta', 'sum')
)

venta_mes
```

```python
# Top 10 productos más vendidos
productos_cantidad_venta = df_ventas.groupby('NombreProducto', as_index = False).agg(
    Quantity = ('Quantity', 'sum')).sort_values('Quantity', ascending = False).head(10)

productos_cantidad_venta
```

```python
# Ingreso promedio por venta
promedio_por_venta = df_ventas['TotalVenta'].mean().round(2)

print(f'Ingreso promedio es: {promedio_por_venta}')
```
