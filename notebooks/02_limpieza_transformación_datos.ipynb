```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_ventas = pd.read_csv('../dataset/raw/ventas.csv')
df_ventas.head()
```

```python
# Correccción de tipo de datos 
df_ventas['fecha'] = pd.to_datetime(df_ventas['fecha'])

df_ventas['Mes'] = df_ventas['fecha'].dt.to_period('M')

df_ventas.info()
```

```python
# Creamos una columna CategoriaVentas
df_ventas["CategoriaVentas"] = pd.cut(df_ventas["TotalVenta"], 
                               bins=[0, 100, 1000, 5000, 16000], 
                               labels=["Baja", "Media", "Alta", "Muy Alta"])

df_ventas.head(60)
```

```python
# Validación final
df_ventas.info()
```
