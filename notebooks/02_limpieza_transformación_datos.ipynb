```python
import pandas as pd
import sys
sys.path.append('..')
from python.c_limpieza.limpieza_datos import *
```

```python
df_ventas = pd.read_csv('../dataset/raw/ventas.csv')
df_ventas.head()
```

```python
# Correccción de tipo de datos 
#f_convertir_fechas(df_ventas, 'fecha')
f_convertir_fechas(df_ventas, 'fecha')

df_ventas.info()
```

```python
# Creación de nueva columna (Mes)
df_ventas['Mes'] = df_ventas['fecha'].dt.to_period('M').astype(str)
df_ventas.head()
```

```python
# Creamos una columna CategoriaVentas
df_ventas["CategoriaVentas"] = pd.cut(df_ventas["TotalVenta"], 
                               bins=[0, 100, 1000, 5000, 16000], 
                               labels=["Baja", "Media", "Alta", "Muy Alta"])

df_ventas.head()
```
